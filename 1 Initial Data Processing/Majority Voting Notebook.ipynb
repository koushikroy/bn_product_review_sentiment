# Majority Voting
Project: **Bangla Product Review Sentiment Analysis**

Objectives of this notebook:
- Take the initial dataset labeled by three different individual
- Use majority voting and combine their labels

## Load Dataset

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel("data/data_1.xlsx")
df2 = pd.read_excel("data/data_2.xlsx")
df3 = pd.read_excel("data/data_3.xlsx")
```

## Merge with Voting and Remove Duplicate

```python
df1.drop_duplicates(inplace = True)
df1.dropna(inplace = True)
df2.drop_duplicates(inplace = True)
df2.dropna(inplace = True)
df3.drop_duplicates(inplace = True)
df3.dropna(inplace = True)

df = pd.concat([df1, df2, df3])
```

```python
def getmode(v):
    uniqv, counts = np.unique(v, return_counts=True)
    return uniqv[np.argmax(counts)]

def sentiment_combined(x):
    return list(x)

df_mod = df.groupby('review', as_index=False).agg(sentiment_combined=('sentiment', sentiment_combined), 
                                                  sentiment_majority=('sentiment', getmode))
df_mod['sentiment_combined'] = df_mod['sentiment_combined'].apply(lambda x: ', '.join(x))
df_mod['sentiment_numeric'] = df_mod['sentiment_majority'].map({'positive': 0, 'negative': 1, 'neutral': 2})
```
