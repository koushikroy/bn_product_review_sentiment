```python
import pandas as pd
from transformers import BertTokenizer, TFBertModel
from tqdm import tqdm

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
bert_multilingual = TFBertModel.from_pretrained("bert-base-multilingual-cased")
```

```python
# Function to get pooled embeddings
count = 0
def get_pooled_embeddings(text):
    global count
    count += 1
    if (count % 10 == 0):
        print(count)
    encoded_input = tokenizer(text, return_tensors='tf')
    output = bert_multilingual(encoded_input)
    pooled_output = output.pooler_output
    return pooled_output.numpy()

# Read xlsx file
file_name = "/content/drive/MyDrive/project/uiu_project/data_analytics_project/word_to_vec/data.xlsx"
df = pd.read_excel(file_name)

# Initialize new column for embeddings
df['embedding'] = None

# Get embeddings for all reviews using tqdm progress bar
for index, row in tqdm(df.iterrows(), total=df.shape[0], desc="Processing reviews"):
    embedding = get_pooled_embeddings(row['review'])
    df.at[index, 'embedding'] = str(embedding.tolist())

# Save the DataFrame to a new xlsx file
df.to_excel("data_with_embeddings.xlsx", index=False)
```
